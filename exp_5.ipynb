**DATA VISUALIZATION USING MATPLOTLIB LIBRARY PART-1**



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**1. Accessing Labels and Line Markers in Line Plots**

```python
# Sample data for line plots
x = range(0, 91, 10)
y = [i + 5 for i in x]  # Data for 2022

z = [i + 10 for i in x]  # Data for 2023


# Sample data for count plot
data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Values': np.random.rand(100)
}
df = pd.DataFrame(data)
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot data with labels and markers
line1, = plt.plot(x, y, label='2022', color='b', marker='o', linestyle='-')
line2, = plt.plot(x, z, label='2023', color='r', marker='x', linestyle='--')

# Add title and labels
plt.title('Line Plot with Custom Markers')
plt.xlabel('First Value')
plt.ylabel('Second Value')

# Add legend
legend = plt.legend()

# Accessing legend labels
print('Legend Labels:')
for label in legend.get_texts():
    print(label.get_text())

# Accessing line markers
print('Line 1 Marker:', line1.get_marker())
print('Line 2 Marker:', line2.get_marker())

# Display the plot
plt.grid(True)
plt.xticks(range(0, 91, 5))
plt.yticks(range(0, 91, 5))
plt.xlim(0, 90)
plt.ylim(0, 90)
plt.show()
```

```python
import seaborn as sns

# Create a count plot
plt.figure(figsize=(8, 6))
count_plot = sns.countplot(data=df, x='Category', palette='viridis')

# Add title and labels
plt.title('Count Plot of Categories')
plt.xlabel('Category')
plt.ylabel('Count')

# Display the count plot
plt.show()
```
