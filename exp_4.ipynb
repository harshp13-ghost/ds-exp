```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.utils import resample
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df
```

**ENCODING**

```python
# Encode the categorical variable (class)
df['Species'] = df['Species'].astype('category').cat.codes
```

```python
df['Species']
```

**NORMALIZATION**

```python
# Separate characteristics and target value
X = df.drop('Species', axis=1)
y = df['Species']

# Normalize the data
min_max_scaler = MinMaxScaler()
X_normalized = min_max_scaler.fit_transform(X)

# Convert back to DataFrame
X_normalized = pd.DataFrame(X_normalized, columns=X.columns)
print(X_normalized.head())
```

**SCALING**

```python
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print(X_scaled.head())
```

**Balancing the Dataset**


```python
# Combine the characteristics and target value
df_scaled = pd.concat([X_scaled, y], axis=1)

# Separate majority and minority classes
df_majority = df_scaled[df_scaled['Species'] == 0]
df_minority = df_scaled[df_scaled['Species'] != 0]

# Upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True, # sample with replacement
                                 n_samples=len(df_majority), # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df_balanced = pd.concat([df_majority, df_minority_upsampled])

# Display the new class distribution
print(df_balanced['Species'].value_counts())
```
