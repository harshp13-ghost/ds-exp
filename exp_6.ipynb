```python
df1 = pd.read_csv('Social_Network_Ads.csv')
df1
```

**Scatter Plot**

```python
plt.figure(figsize = (16,9))
plt.title('Scatter plot')         #title for the plot
plt.xlabel('Age')                 #label for x axis
plt.ylabel('Salary')              #label for y axis
plt.grid()                        #grid for the plot
plt.scatter(df1['Age'], df1['EstimatedSalary'], c = df1['Purchased'])
```

**Bar Plot**

```python
plt.figure(figsize = (16,9))
plt.grid()                        #grid for the plot
plt.xlabel('Age')                 #label for x axis
plt.ylabel('EstimatedSalary')     #label for y axis
plt.title('Bar plot')             #title for the plot
plt.bar(df1['Age'], df1['EstimatedSalary'],  color = 'purple')
```

**Histogram**

```python
plt.xlabel('Age')                 #label for x axis
plt.ylabel('EstimatedSalary')     #label for y axis
plt.title('Histogram')            #title for the plot

plt.grid()                        #grid for the plot
plt.xticks(range(10,61,10))       #ticks for x axis
plt.hist(df1['Age'], bins = range(10,61,10), color ='k' );
```

**Box Plot**

```python
plt.boxplot(df1['Age']);
```
